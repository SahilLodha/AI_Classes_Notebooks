# Progressions

1. Arithmetic
x1, x2, x3, x4 are in AP in case x1 - x2 = x2 - x3 = x3 - x4 and so on.

2. Geometric
x1, x2, x3, x4 are in GP if x1/x2 = x2/x3 = x3/x4 and so on.

3. Harmonic
x1, x2, x3, x4 are in HP if x1^-1 - X2^-1 = X2^-1 - X3^-1 and so on
