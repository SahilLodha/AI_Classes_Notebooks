```python
slices = [120, 80, 30, 60, 1]
labels = ['JS', 'Python', 'C#', 'Scala', 'Go']
colors = ['red', 'blue', 'green', 'orange', 'violet']
```

```python
plt.xkcd()
plt.pie(slices, labels = labels,autopct= '%1.2f%%',colors=colors,wedgeprops={'edgecolor': 'red'})
plt.title('Pie Chart Demonstration')
plt.show()
```
