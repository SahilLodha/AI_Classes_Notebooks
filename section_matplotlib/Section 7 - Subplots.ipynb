```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
plt.rc('figure', figsize=(5, 10))
```

```python
df = pd.read_csv(r'../data/data_10.csv')
df.head(10)
```

```python
ages = df.Age
sal_others = df.All_Devs
sal_py = df.Python
sal_js = df.JavaScript
```

```python
plt.plot(ages, sal_py, label='Python', color='red')
plt.plot(ages, sal_js, label='JS', color='green')
plt.legend()
plt.grid(True)
```

```python
fig, ax = plt.subplots()
ax.plot(ages, sal_py, label='Python', color='red')
ax.plot(ages, sal_js, label='JS', color='green')
ax.legend()
ax.set_xlabel('Ages')
ax.set_ylabel('Salary per Year in USD.')
ax.grid(True)
plt.tight_layout()
```

```python
fig, axis = plt.subplots(2, 2)

axis[0][0].plot(ages, sal_py, label='Python', color='red')
axis[1][0].pie([sal_js, sal_py, sal_others], labels=['JS', 'Python', 'others'])
axis[0][1].plot(ages, sal_others, label='Others', color='yellow')
'''
ax1[0].legend()
ax2[0].legend()
ax1[1].legend()

ax1[0].set_xlabel('Ages')
ax1[0].set_ylabel('Salary per Year in USD')

ax2[0].set_xlabel('Ages')
ax2[0].set_ylabel('Salary per Year in USD')

ax1[1].set_xlabel('Ages')
ax1[1].set_ylabel('Salary per Year in USD')
'''

plt.tight_layout()
```
