```python
from matplotlib import pyplot as plt
import numpy as np
import csv
from collections import Counter
```

```python
age_X = [25, 30, 35, 40, 45, 50,  55, 60, 65]
avg_SAL_NP = [25000,25300, 45000,54500, 58000, 68000, 54000, 20000,30000]
avg_SAL_US = [3500,2300, 4800,5400, 5000, 6700, 5400, 2000,3000]

plt.bar(age_X, avg_SAL_NP, color='#444444' ,label="Salaries-NP", width=3)
plt.bar(age_X, avg_SAL_US, color='yellow' ,label="Salaries-US", width=3)
plt.title("Stacked Bar Chart")
plt.legend()
```

```python
year = [i for i in range(2005, 2010)]

itsutra_rev = np.random.randint(1000, 5000, len(year))
lisnepal_rev = np.random.randint(1000, 5000, len(year))
cedargate_rev = np.random.randint(5000, 7500, len(year))

x_index = np.arange(len(year))
bar_width = 0.5

plt.bar(x_index-bar_width, itsutra_rev, width=bar_width/3, color='blue', label='IT sutra')
plt.bar(x_index, lisnepal_rev, width=bar_width/3, color='green', label='LIS Nepal')
plt.bar(x_index+bar_width, cedargate_rev, width=bar_width/3, color='orange', label='Cedar Gate')      

plt.xticks(ticks=x_index, labels=year)
plt.legend()
```

```python
## opening data file
with open(r'../data/data.csv') as f:
    csv_reader = csv.DictReader(f)
    language_counter = Counter()

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

language_counter

languages = []
popularity = []

for lan, pop in language_counter.most_common(20):
    languages.append(lan)
    popularity.append(pop)

languages.reverse()
popularity.reverse()
```

```python
plt.barh(languages, popularity)
plt.xlabel("Count of Users")
plt.ylabel("Languages")
plt.title("Language VS Active Users")
```
