```python
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# List all the styles Present in the matplotlib
# plt.style.available 

plt.style.use('seaborn-v0_8-white')
plt.rc('figure', figsize=(12, 10))
```

```python
df = pd.read_csv('../data/timeseries_demo.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', inplace=True)

df.to_csv('../file_io/timeseries.csv')
```

```python
price_date = df.Date
price_closing = df.Close
price_low = df.Low
mean_closing = price_low.mean()
mean_closing
```

```python
plt.xkcd()
plt.plot_date(price_date, price_closing, linestyle='solid')
plt.plot_date(price_date, price_low, linestyle=':', color='g')

plt.fill_between(price_date, price_closing, alpha=0.25, y2=mean_closing, color='green', where=(price_closing > mean_closing), interpolate=True)
plt.fill_between(price_date, price_closing, alpha=0.25, y2=mean_closing, color='red', where=(price_closing <= mean_closing), interpolate=True)

plt.gcf().autofmt_xdate()
plt.grid(True)
```
