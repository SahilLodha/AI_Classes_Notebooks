```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.random.randint(1, 20, 40)
y = np.random.randint(1, 10, 40)
sizes = np.random.randint(5, 100, 40)

plt.xkcd()
plt.scatter(x, y, s=sizes, c=sizes, cmap='Reds', edgecolors='black', alpha=0.9, linewidth=1)
cbar = plt.colorbar()
cbar.set_label("Salary Level")
plt.tight_layout()
```
