```python
import numpy
import matplotlib.pyplot as plt
```

```python
ages = numpy.random.randint(0, 99, 10000)
```

```python
ages.max()
mean_age = ages.mean()
median_age = numpy.median(ages)
mean_age, median_age
```

```python
bins = [i for i in range(0, 100, 9)]
labels = [f'{i}-{i+9}' for i in bins]
print(labels)
```

```python
plt.hist(ages, bins=bins, edgecolor='black', label=labels)
plt.title("Distribution of Age")
plt.axvline(mean_age, color='r', linestyle=':')
# plt.axvline(median_age, color='g')
plt.show()
```
